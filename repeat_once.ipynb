```python
%%sql
--use a cte
with customer_order as(
    select customerkey, count(*) as total_orders from sales 
    group by customerkey
)
select 
    case
        when total_orders = 1 then 'once'
        else 'Repeat_cutomer'
    end as type_of_customer,
    count(*) as num_customers,
    sum(quantity*netprice*exchangerate) as revenue
from customer_order co
left join sales s on co.customerkey = s.customerkey
group by type_of_customer

```
