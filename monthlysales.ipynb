we want to find find sales over a particular period

```python
%%sql
select extract(MONTH from orderdate) as month, p.categoryname, sum(quantity*unitprice*exchangerate)
from sales s
left join product p on s.productkey = p.productkey
where extract(YEAR from orderdate) = 2023
group by month,p.categoryname
```

according to gemini there is a dip in sales in the 4th month

lets check yearly

```python
%%sql
select extract(year from orderdate) as year, p.categoryname, sum(quantity*unitprice*exchangerate)
from sales s
left join product p on s.productkey = p.productkey
group by year,p.categoryname
```

```python
%%sql
select TO_CHAR(orderdate, 'YYYY-MM') as month, p.categoryname, sum(quantity*unitprice*exchangerate)
from sales s
left join product p on s.productkey = p.productkey
where extract(YEAR from orderdate) = 2023
group by month,p.categoryname
```
