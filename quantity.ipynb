```python
import config

connection_string = f"postgresql://{config.DB_USER}:{config.DB_PASSWORD}@{config.DB_HOST}:{config.DB_PORT}/{config.DB_NAME}"

%reload_ext sql
%sql $connection_string
```

```python
%%sql
select p.categoryname, sum(quantity) as quantity
from sales s
left join product p on s.productkey = p.productkey
group by p.categoryname
order by quantity desc
```

```python
%%sql
select p.categoryname, sum(quantity*unitprice*exchangerate) as revenue
from sales s
left join product p on s.productkey = p.productkey
group by p.categoryname
order by revenue desc
```
