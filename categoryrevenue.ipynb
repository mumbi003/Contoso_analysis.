join with the sales table to get the revenue 

```python
%%sql
select * from sales limit 10
```

```python
%%sql
select p.categoryname, round(sum(quantity * unitprice*exchangerate)) as sales from sales s
left join product p on s.productkey = p.productkey
group by p.categoryname
order by p.categoryname
```
